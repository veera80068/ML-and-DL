### 1. Define probability. How is it calculated in simple events?

**Answer:** 
- **Definition:** Probability is a measure of the likelihood that a particular event will occur. It ranges from 0 (impossible event) to 1 (certain event).
  
- **Calculation in Simple Events:** In simple events (equally likely outcomes), the probability of an event (P(A)) is calculated using the formula: 
  \[ P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}} \]

---

### 2. What is a probability distribution and why is it important?

**Answer:** 
- **Definition:** A probability distribution describes the likelihood of different outcomes in a sample space. It provides the probabilities for all possible outcomes of a random variable.
  
- **Importance:** Probability distributions are fundamental in statistics for modeling uncertainty. They allow us to make predictions, calculate expected values, and assess the likelihood of various events.

---

### 3. Explain the differences between discrete and continuous probability distributions.

**Answer:** 
- **Discrete Probability Distribution:** Deals with countable outcomes and assigns probabilities to distinct, separate values. Examples include the binomial and Poisson distributions.
  
- **Continuous Probability Distribution:** Deals with uncountable outcomes, often in a range. It assigns probabilities to intervals and requires integration for calculations. Examples include the normal and exponential distributions.

---

### 4. Define the Bernoulli distribution and provide an example of its application.

**Answer:** 
- **Definition:** The Bernoulli distribution models a single experiment with two possible outcomes: success (usually denoted as 1) and failure (denoted as 0).
  
- **Example:** A coin flip, where success might be getting heads (1) and failure getting tails (0).

---

### 5. What is the difference between a binomial and a multinomial distribution?

**Answer:** 
- **Binomial Distribution:** Deals with a fixed number of independent trials, each with two possible outcomes (success or failure).
  
- **Multinomial Distribution:** Deals with experiments involving multiple categories or outcomes, each with its own probability.

---

### 6. Define the Poisson distribution and give an example of its use in real-world scenarios.

**Answer:** 
- **Definition:** The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a known average rate.
  
- **Example:** The number of emails received in an hour, the number of calls at a call center in a minute.

---

### 7. Explain the concept of a normal distribution. What are its key characteristics?

**Answer:** 
- **Concept:** The normal distribution, also known as the Gaussian distribution, is symmetric and bell-shaped. It is characterized by its mean (center) and standard deviation (spread).
  
- **Key Characteristics:** 
  - About 68% of data falls within one standard deviation of the mean.
  - About 95% falls within two standard deviations.
  - About 99.7% falls within three standard deviations.

---

### 8. Define the exponential distribution and provide an example of its application.

**Answer:** 
- **Definition:** The exponential distribution models the time between independent events occurring at a constant rate.
  
- **Example:** The time between arrivals of buses at a bus stop, the time between customer arrivals at a service point.

---

### 9. What is the central limit theorem and why is it important in statistics?

**Answer:** 
- **Central Limit Theorem (CLT):** The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original distribution.
  
- **Importance:** It allows statisticians to make inferences about a population mean, even if the population distribution is not normal. The CLT is fundamental in hypothesis testing and estimation.

---

### 10. Explain the differences between a uniform, normal, and exponential distribution.

**Answer:** 
- **Uniform Distribution:** All values in the range are equally likely. Probability density is constant.
  
- **Normal Distribution:** Symmetric, bell-shaped curve with a defined mean and standard deviation. Most values cluster around the mean.
  
- **Exponential Distribution:** Skewed to the right, models the time between events in a Poisson process. Probability of longer times decreases exponentially.
